### Task 1: Visualizing Data Distributions


##### 1. Histogram of `ConvertedCompYearly`


Visualize the distribution of yearly compensation (`ConvertedCompYearly`) using a histogram.



```python
comp = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce').dropna()

upper_limit = comp.quantile(0.95)

plt.figure(figsize=(8,6))
plt.hist(comp[comp <= upper_limit], bins=30)
plt.title('Distribution of Yearly Compensation (ConvertedCompYearly)')
plt.xlabel('ConvertedCompYearly')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

##### 2. Box Plot of `Age`


Since `Age` is categorical in the dataset, convert it to numerical values for a box plot.



```python
import numpy as np 

age_map = {'Under 18 years old':1, '18-24 years old':2, '25-34 years old':3, '35-44 years old':4,
           '45-54 years old':5, '55-64 years old':6, '65 years or older':7, 'Prefer not to say':np.nan}

df['AgeNumeric'] = df['Age'].map(age_map)
comp = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')

plt.figure(figsize=(8,6))
plt.boxplot(
    [comp[df['AgeNumeric'] == i].dropna() for i in sorted(df['AgeNumeric'].dropna().unique())],
    showfliers=False
)
plt.xticks(range(1, 8), ['Under 18','18-24','25-34','35-44','45-54','55-64','65+'], rotation=45)
plt.ylim(bottom=0)
plt.title('Yearly Compensation by Age Group')
plt.ylabel('ConvertedCompYearly')
plt.tight_layout()
plt.show()
```

### Task 2: Visualizing Relationships in Data


##### 1. Scatter Plot of `Age_numeric` and `ConvertedCompYearly`


Explore the relationship between age and compensation.



```python
age_map = {'Under 18 years old':1, '18-24 years old':2, '25-34 years old':3, '35-44 years old':4,
           '45-54 years old':5, '55-64 years old':6, '65 years or older':7, 'Prefer not to say':np.nan}

df['AgeNumeric'] = df['Age'].map(age_map)
comp = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')

plt.figure(figsize=(8,6))
plt.scatter(df['AgeNumeric'], comp, alpha=0.5)

plt.title('Age vs Yearly Compensation')
plt.xlabel('Age Group (Numeric)')
plt.ylabel('ConvertedCompYearly')
plt.ylim(0, 600000)
plt.tight_layout()
plt.show()
```

##### 2. Bubble Plot of `ConvertedCompYearly` and `JobSatPoints_6` with `Age_numeric` as Bubble Size


Explore how compensation and job satisfaction are related, with age as the bubble size.


```python
age_map = {'Under 18 years old':1,'18-24 years old':2,'25-34 years old':3,'35-44 years old':4,
           '45-54 years old':5,'55-64 years old':6,'65 years or older':7,'Prefer not to say':np.nan}

df['AgeNumeric'] = df['Age'].map(age_map)
comp = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')
jobsat = pd.to_numeric(df['JobSat'], errors='coerce')

data = pd.DataFrame({'comp': comp, 'jobsat': jobsat, 'age': df['AgeNumeric']}).dropna()

plt.figure(figsize=(8,6))
plt.scatter(data['comp'], data['jobsat'], s=data['age']*30, alpha=0.5)

plt.title('Compensation vs Job Satisfaction (Bubble Size = Age)')
plt.xlabel('ConvertedCompYearly')
plt.ylabel('JobSat')
plt.xlim(0, 200000)
plt.tight_layout()
plt.show()
```

### Task 3: Visualizing Composition of Data with Bar Charts


##### 1. Horizontal Bar Chart of `MainBranch` Distribution


Visualize the distribution of respondents’ primary roles to understand their professional focus.



```python
mainbranch = df['MainBranch'].value_counts()

plt.figure(figsize=(10,6))
plt.barh(mainbranch.index, mainbranch.values)
plt.title('Distribution of MainBranch')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.tight_layout()
plt.show()
```

##### 2. Vertical Bar Chart of Top 5 Programming Languages Respondents Want to Work With


Identify the most desired programming languages based on `LanguageWantToWorkWith`.



```python
lang = df['LanguageWantToWorkWith'].dropna().str.split(';').explode()
top_lang = lang.value_counts().head(5)

plt.figure(figsize=(8,6))
plt.bar(top_lang.index, top_lang.values)
plt.title('Top 5 Programming Languages Respondents Want to Work With')
plt.xlabel('Programming Language')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##### 3. Stacked Bar Chart of Median `JobSatPoints_6` and `JobSatPoints_7` by Age Group


Compare job satisfaction metrics across different age groups with a stacked bar chart.


```python
sat = df[['Age','JobSatPoints_6','JobSatPoints_7']].copy()

sat['JobSatPoints_6'] = pd.to_numeric(sat['JobSatPoints_6'], errors='coerce')
sat['JobSatPoints_7'] = pd.to_numeric(sat['JobSatPoints_7'], errors='coerce')

median_sat = sat.groupby('Age')[['JobSatPoints_6','JobSatPoints_7']].median()
median_sat.plot(kind='bar', stacked=True, figsize=(10,6))

plt.title('Median Job Satisfaction (Points 6 & 7) by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##### 4. Bar Chart of Database Popularity (`DatabaseHaveWorkedWith`)


Identify the most commonly used databases among respondents by visualizing `DatabaseHaveWorkedWith`.



```python
db = df['DatabaseHaveWorkedWith'].dropna().str.split(';').explode()
top_db = db.value_counts().head(10)

plt.figure(figsize=(8,6))
plt.bar(top_db.index, top_db.values)
plt.title('Most Common Databases Used by Respondents')
plt.xlabel('Database')
plt.ylabel('Count')
plt.xticks(range(len(top_db)), top_db.index, rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Comparison of Data with Bar Charts


##### 1. Grouped Bar Chart of Median `ConvertedCompYearly` for Different Age Groups


Compare median compensation across multiple age groups with a grouped bar chart.



```python
comp_age = df[['Age', 'ConvertedCompYearly']].dropna().copy()
comp_age['ConvertedCompYearly'] = pd.to_numeric(comp_age['ConvertedCompYearly'], errors='coerce')

median_comp = comp_age.groupby('Age')['ConvertedCompYearly'].median()

plt.figure(figsize=(10,6))
plt.bar(range(len(median_comp)), median_comp.values)
plt.title('Median Yearly Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median ConvertedCompYearly')
plt.xticks(range(len(median_comp)), median_comp.index, rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

##### 2. Bar Chart of Respondent Count by Country


Show the distribution of respondents by country to see which regions are most represented.



```python
country = df['Country'].value_counts().head(10)

plt.figure(figsize=(10,6))
plt.bar(range(len(country)), country.values)

plt.title('Top Countries by Respondent Count')
plt.xlabel('Country')
plt.ylabel('Number of Respondents')
plt.xticks(range(len(country)), country.index, rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
