## Identify and Analyze Duplicates


### Task 1: Identify Duplicate Rows
1. Count the number of duplicate rows in the dataset.
3. Display the first few duplicate rows to understand their structure.


```python
## Duplicate number of rows
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)
```

```python
## First few duplicate rows
duplicate_rows = df[df.duplicated()]
print("First few duplicate rows:")
print(duplicate_rows.head())
```

### Task 2: Analysis of Duplicate Response Patterns
1. Identify duplicate response patterns based on selected columns such as MainBranch, Employment, and RemoteWork.
2. Clarify that these represent multiple respondents with identical answers rather than duplicate records. Analyse which other columns frequently share identical values within these response-pattern groups.
   


```python
## Write your code here
# Select key columns
key_columns = ['MainBranch', 'Employment', 'RemoteWork']

# Group by these columns to identify shared response patterns
pattern_counts = df.groupby(key_columns).size().reset_index(name='Count')

# Show patterns where more than one respondent has the same responses
duplicate_patterns = pattern_counts[pattern_counts['Count'] > 1]

print("Duplicate response patterns:")
print(duplicate_patterns.head())

# Analyse which other columns share identical values in these groups
grouped = df.groupby(key_columns)

# Check columns that have identical values within groups
for column in df.columns:
    if column not in key_columns:
        identical = grouped[column].nunique()
        if (identical == 1).sum() > 0:
            print(f"Column '{column}' often has identical values within response groups")
```

### Task 3: Visualize Shared Response Patterns
1. Create visualizations to show the distribution of shared response patterns across different categories.
2. Use bar charts or pie charts to represent the distribution of respondents who share identical values for MainBranch, Employment, and RemoteWork, grouped by Country and Employment.


```python
# Group data by Country and Employment for respondents sharing response patterns
pattern_distribution = df.groupby(['Country', 'Employment']).size().reset_index(name='Count')

# Bar chart for distribution
plt.figure(figsize=(10,6))
pattern_distribution.sort_values('Count', ascending=False).head(10).plot(
    kind='bar',
    x='Country',
    y='Count',
    legend=False
)

plt.title("Distribution of Shared Response Patterns by Country")
plt.ylabel("Number of Respondents")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.show()


# Pie chart for Employment distribution
employment_counts = df['Employment'].value_counts()

plt.figure(figsize=(8,8))
employment_counts.head(5).plot(kind='pie', autopct='%1.1f%%')

plt.title("Distribution of Employment Types")
plt.ylabel("")
plt.show()
```
