Print the mean age of the survey participants.


```python
## Write your code here
df['Age'].unique() # values in the age column is in string
```

```python
import numpy as np

df_new = df.copy()

# Initialize numeric column with NaN
df_new['age_numeric'] = np.nan

# Handle "Under 18" → 17
df_new.loc[df_new['Age'].str.contains('under 18', case=False, na=False), 'age_numeric'] = 17

# Handle "65 years or older" → 65
df_new.loc[df_new['Age'].str.contains('65 years or older', case=False, na=False), 'age_numeric'] = 65

# Handle ranges like "18-24" → midpoint
range_mask = df_new['Age'].str.contains(r'\d+\s*-\s*\d+', regex=True, na=False)
ranges = df_new.loc[range_mask, 'Age'].str.extract(r'(\d+)\s*-\s*(\d+)').astype(float)
df_new.loc[range_mask, 'age_numeric'] = (ranges[0] + ranges[1]) / 2

# "Prefer not to say" → leave as NaN automatically

# Ensure the numeric column is float
df_new['age_numeric'] = df_new['age_numeric'].astype(float)

# Now df_new has all the original columns plus 'age_numeric'
mean_age = df_new['age_numeric'].mean()

print(df_new)
print("Mean age:", mean_age)
```
