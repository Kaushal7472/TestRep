**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Read the CSV file
df = pd.read_csv('survey-data.csv')

# Display the first few rows of the data
df.head()
```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
## Write your code here
query = "SELECT CompTotal FROM main WHERE CompTotal IS NOT NULL"
df = pd.read_sql_query(query, conn)

df["CompTotal"] = pd.to_numeric(df["CompTotal"], errors="coerce")
comp = df["CompTotal"].dropna()
comp = comp[comp > 0]

plt.hist(comp[comp < 500000], bins=50)
plt.title("Distribution of Total Compensation")
plt.xlabel("CompTotal")
plt.ylabel("Frequency")
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
QUERY = "SELECT Age FROM main WHERE Age IS NOT NULL"
df = pd.read_sql_query(QUERY, conn)

print(df["Age"].drop_duplicates().tolist())

age_map = {
    "Under 18 years old": 17,
    "18-24 years old": 21,
    "25-34 years old": 29.5,
    "35-44 years old": 39.5,
    "45-54 years old": 49.5,
    "55-64 years old": 59.5,
    "65 years or older": 65
}

df["AgeNumeric"] = df["Age"].map(age_map)

print("Rows after mapping:", df["AgeNumeric"].notna().sum())


plt.figure(figsize=(6,4))
plt.boxplot(df["AgeNumeric"].dropna())
plt.title("Box Plot of Age")
plt.ylabel("Approximate Age")
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
query = """
SELECT Age, WorkExp
FROM main
WHERE Age IS NOT NULL AND WorkExp IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# Map Age categories to approximate numeric values
age_map = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 65
}

df["Age"] = df["Age"].map(age_map)
df["WorkExp"] = pd.to_numeric(df["WorkExp"], errors="coerce")

df = df.dropna()

print(df.head())
print("Rows left:", len(df))

plt.figure(figsize=(8,5))
plt.scatter(df["Age"], df["WorkExp"])
plt.xlabel("Age")
plt.ylabel("Work Experience")
plt.title("Age vs Work Experience")
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
query = """
SELECT Age, TimeSearching, Frustration
FROM main
WHERE Age IS NOT NULL
AND TimeSearching IS NOT NULL
AND Frustration IS NOT NULL
"""

df = pd.read_sql_query(query, conn)

# convert frustration selections into a count
df["FrustrationCount"] = df["Frustration"].apply(lambda x: len(x.split(";")))

age_map = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 65,
    'Prefer not to say': None
}

time_map = {
    'Less than 15 minutes a day': 1,
    '15-30 minutes a day': 2,
    '30-60 minutes a day': 3,
    '60-120 minutes a day': 4,
    'Over 120 minutes a day': 5
}

df["AgeNum"] = df["Age"].map(age_map)
df["TimeSearchingNum"] = df["TimeSearching"].map(time_map)
df = df.dropna()

plt.figure(figsize=(8,5))
plt.scatter(df["TimeSearchingNum"],df["FrustrationCount"],s=df["AgeNum"]*3,alpha=0.5)
plt.xlabel("Time Searching")
plt.ylabel("Number of Frustrations Selected")
plt.title("Bubble Plot: Time Searching vs Frustration")
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
query = """
SELECT DatabaseWantToWorkWith
FROM main
WHERE DatabaseWantToWorkWith IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# Split multi-select values into separate database names
db_series = df["DatabaseWantToWorkWith"].str.split(";").explode()

# Count top 5
top5 = db_series.value_counts().head(5)

plt.figure(figsize=(7,7))
plt.pie(top5.values, labels=top5.index, autopct='%1.1f%%', startangle=90)
plt.title("Top 5 Databases Respondents Want to Learn Next Year")
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
query = """
SELECT Age, TimeSearching, TimeAnswering
FROM main
WHERE Age IS NOT NULL
  AND TimeSearching IS NOT NULL
  AND TimeAnswering IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

time_map = {
    'Less than 15 minutes a day': 1,
    '15-30 minutes a day': 2,
    '30-60 minutes a day': 3,
    '60-120 minutes a day': 4,
    'Over 120 minutes a day': 5
}

df["TimeSearchingNum"] = df["TimeSearching"].map(time_map)
df["TimeAnsweringNum"] = df["TimeAnswering"].map(time_map)

# nearest age group for 30 to 35
df_filtered = df[df["Age"] == "25-34 years old"]

df_filtered = df_filtered.dropna(subset=["TimeSearchingNum", "TimeAnsweringNum"])

median_searching = df_filtered["TimeSearchingNum"].median()
median_answering = df_filtered["TimeAnsweringNum"].median()

plot_df = pd.DataFrame({
    "TimeSearching": [median_searching],
    "TimeAnswering": [median_answering]
}, index=["25-34 years old"])

plot_df.plot(kind="bar", stacked=True)

plt.title("Median TimeSearching and TimeAnswering")
plt.ylabel("Median Category Value")
plt.xlabel("Age Group")
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
## Write your code here
query = """
SELECT Age, CompTotal
FROM main
WHERE Age IS NOT NULL
  AND CompTotal IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

age_map = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 65,
    'Prefer not to say': None
}

df["AgeNum"] = df["Age"].map(age_map)
df["CompTotal"] = pd.to_numeric(df["CompTotal"], errors="coerce")

df = df.dropna(subset=["AgeNum", "CompTotal"])

# keep ages from 45 to 60
df_filtered = df[(df["AgeNum"] >= 45) & (df["AgeNum"] <= 60)]

median_comp = df_filtered.groupby("AgeNum")["CompTotal"].median()

plt.figure(figsize=(8,5))
plt.plot(median_comp.index, median_comp.values, marker='o')
plt.xlabel("Age")
plt.ylabel("Median CompTotal")
plt.title("Median CompTotal for Ages 45 to 60")
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
## Write your code here
query = """
SELECT MainBranch
FROM main
WHERE MainBranch IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# Count occurrences
branch_counts = df["MainBranch"].value_counts()

plt.figure(figsize=(8,5))
branch_counts.plot(kind="barh")

plt.xlabel("Number of Respondents")
plt.ylabel("Main Branch")
plt.title("Respondents by Main Branch")
plt.show()
```
